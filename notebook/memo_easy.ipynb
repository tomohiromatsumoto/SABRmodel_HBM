```python
import numpy
import pylab
import math
import time 

X = 0 # 的に当たった回数
N = 500 # 試行回数

# N回の試行を開始
for i in range(1, N):
    score_x = numpy.random.rand()
    score_y = numpy.random.rand()
    if score_x * score_x + score_y * score_y < 1:
        X = X + 1
# piの近似値をここで計算する
    pi = 4*float(X)/float(i)
    plot_x = i
    plot_y = pi
    pylab.plot(plot_x,plot_y,"ro")

# 結果を表示 
pylab.grid(True)
pylab.xlabel('X')
pylab.ylabel('Y')
pylab.show()

print(X)
print(500 - X)
```
